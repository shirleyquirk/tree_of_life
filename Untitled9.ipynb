```python
import math
#rendering with threads, oh god not ever going to work.
dwg=svgwrite.Drawing(size=(500,500))
def cheaptaperedpath(start,theta0,theta1,end,startwidth,endwidth):
    norm0=Vec2D(r=startwidth/2,t=theta0+math.pi/2)
    norm1=Vec2D(r=endwidth/2,t=theta1+math.pi/2)
    p0_0=start+norm0
    p0_1=start-norm0
    p3_0=end+norm1
    p3_1=end-norm1
    d=end-start
    d0=Vec2D(r=0.45*d.r,t=theta0)
    p1_0=p0_0+d0
    p1_1=p0_1+d0
    d1=Vec2D(r=0.3*d.r,t=theta1+math.pi)
    p2_0=p3_0+d1
    p2_1=p3_1+d1
    return svgwrite.path.Path(d=['M',p0_0.car,'C',p1_0.car,p2_0.car,p3_0.car,'L',p3_1.car,'C',p2_1.car,p1_1.car,p0_1.car],fill='blue',stroke='none',debug=False)
dwg.add(cheaptaperedpath(Vec2D(100,100),radians(90),radians(90),Vec2D(300,300),10,5))
from IPython.display import SVG
SVG(dwg.tostring())
```
